```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

a = np.random.randint(0, 10, size=(3, 3))  # Random integers between 0 and 9
b = np.random.randint(0, 10, size=(3, 3))  # Random integers between 0 and 9

print("Matrix A:\n", a)
print("Matrix B:\n", b)

# Matrix addition
c = a + b
print("\nThe sum of a and b is:\n", c)

# Matrix subtraction
d = a - b
print("\nThe difference of a and b is:\n", d)

# Matrix multiplication
e = np.matmul(a, b)  # or simply 'a @ b'
print("\nThe product of a and b is:\n", e)

#  Determinant of a matrix
det_a = np.linalg.det(a)  # Returns the determinant of the matrix

# Inverse of a matrix
if np.linalg.det(a) != 0:  # Check if matrix is invertible
    inv_a = np.linalg.inv(a)  # Compute the inverse of A
else:
    inv_a = "Matrix is singular, cannot compute inverse"  # No inverse if determinant is 0

# Rank of a matrix
rank_a = np.linalg.matrix_rank(a)  # Returns the rank of the matrix

print("Determinant of A:", det_a)
print("Inverse of A:\n", inv_a)
print("Rank of A:", rank_a)
```

```python
import matplotlib.pyplot as plt

x = np.random.rand(15)
y = np.random.rand(15)

plt.plot(range(15), x, marker='o', label='Vector X')
plt.plot(range(15), y, marker='s', label='Vector Y')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Two Random Vectors')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

matrix_4x4 = np.random.rand(4, 4)

sns.heatmap(matrix_4x4, annot=True, cmap='viridis', linewidths=0.5, linecolor='black')
plt.title("Heatmap of 4x4 Random Matrix")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

```python
A = np.random.randint(1, 10, (4, 4))
B = np.random.randint(1, 10, (4, 4))

sum_matrix = A + B
diff_matrix = A - B
prod_matrix = A @ B

# Flatten matrices for plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot Addition
axes[0].bar(range(16), sum_matrix.flatten(), color='blue')
axes[0].set_title("Matrix Addition")
axes[0].set_xlabel("Element Index")
axes[0].set_ylabel("Value")

# Plot Subtraction
axes[1].bar(range(16), diff_matrix.flatten(), color='green')
axes[1].set_title("Matrix Subtraction")
axes[1].set_xlabel("Element Index")
axes[1].set_ylabel("Value")

# Plot Multiplication
axes[2].bar(range(16), prod_matrix.flatten(), color='red')
axes[2].set_title("Matrix Multiplication")
axes[2].set_xlabel("Element Index")
axes[2].set_ylabel("Value")

plt.tight_layout()
plt.show()
```
